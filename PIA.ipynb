```python
import tensorflow as tf
import numpy as np

# =====================================================
# DATOS DE ENTRADA (X)
# [humedadSuelo, lluvia]
#
# NORMALIZACIГ“N:
# 0.0  -> valor bajo
# 1.0  -> valor alto
#
# SENSOR LLUVIA:
# 0.0 = LLUEVE
# 1.0 = NO LLUEVE
#
# SENSOR HUMEDAD:
# 0.0 = MUY SECO
# 1.0 = MUY HГљMEDO
# =====================================================

X = np.array([

    # seco + NO lluvia
    [0.1, 0.9],

    # seco + lluvia
    [0.1, 0.1],

    # muy hГєmedo + lluvia
    [0.9, 0.1],

    # hГєmedo + poca lluvia
    [0.9, 0.8],

    # humedad media + lluvia
    [0.5, 0.2],

    # muy seco + sin lluvia
    [0.05, 0.95],

    # muy seco + lluvia fuerte
    [0.05, 0.05],

    # humedad correcta + no lluvia
    [0.6, 0.9],

    # seco moderado + no lluvia
    [0.3, 0.9],

    # seco moderado + lluvia
    [0.3, 0.2],

], dtype=np.float32)

# =====================================================
# SALIDAS (Y)
# [bomba, servo]
#
# bomba:
# 1 = ENCENDER
# 0 = APAGAR
#
# servo:
# 1 = ABRIR TECHO
# 0 = CERRAR TECHO
# =====================================================

Y = np.array([

    [1, 0],  # regar
    [0, 1],  # aprovechar lluvia
    [0, 0],  # exceso humedad
    [0, 0],  # no hacer nada
    [0, 1],  # usar lluvia
    [1, 0],  # regar urgente
    [0, 1],  # lluvia fuerte
    [0, 0],  # humedad correcta
    [1, 0],  # necesita riego
    [0, 1],  # aprovechar lluvia

], dtype=np.float32)

# =====================================================
# MODELO DE RED NEURONAL
# =====================================================

model = tf.keras.Sequential([

    tf.keras.layers.Input(shape=(2,)),

    tf.keras.layers.Dense(
        8,
        activation='relu'
    ),

    tf.keras.layers.Dense(
        8,
        activation='relu'
    ),

    tf.keras.layers.Dense(
        2,
        activation='sigmoid'
    )

])

# =====================================================
# COMPILACIГ“N
# =====================================================

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# =====================================================
# ENTRENAMIENTO
# =====================================================

model.fit(
    X,
    Y,
    epochs=300,
    verbose=1
)

# =====================================================
# PRUEBAS
# =====================================================

tests = np.array([

    [0.1, 0.9],  # seco sin lluvia
    [0.1, 0.1],  # seco con lluvia
    [0.9, 0.1],  # muy hГєmedo con lluvia
    [0.4, 0.95], # humedad media sin lluvia

], dtype=np.float32)

predicciones = model.predict(tests)

print("\n==============================")
print("PREDICCIONES")
print("==============================")

for i in range(len(tests)):

    print("\nEntrada:", tests[i])

    bomba = predicciones[i][0]
    servo = predicciones[i][1]

    print("Bomba:", bomba)
    print("Servo:", servo)

# =====================================================
# GUARDAR MODELO KERAS
# =====================================================

model.save("invernadero.keras")

print("\nModelo Keras guardado")

# =====================================================
# CONVERTIR A TFLITE
# =====================================================

converter = tf.lite.TFLiteConverter.from_keras_model(model)

tflite_model = converter.convert()

with open("invernadero.tflite", "wb") as f:
    f.write(tflite_model)

print("Modelo TFLite guardado")
```
